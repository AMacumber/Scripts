# Label Encoding

Many analyses require numerical data. Label encoding transforms categorical variables to numerical variables.

```python
# modules needed for data import
import pandas as pd

# modules needed for Label Encoding
from sklearn import preprocessing
```

```python
# load data
data = pd.read_csv("Data_Churn_HR.csv")
```

```python
data.head()
```

```python
# create instance of labelEncoder
le = preprocessing.LabelEncoder()

# Converting string labels into numbers
data['salary'] = le.fit_transform(data['salary'])
data['Departments '] = le.fit_transform(data['Departments '])
```
